```python
# Example data - replace with your own model's output
y_true = np.array([0, 0, 1, 1])  # True labels
y_score = np.array([0.1, 0.4, 0.35, 0.8])  # Predicted scores for the positive class

# Dummy model (usually random classifier)
dummy_fpr = np.array([0.0, 1.0])
dummy_tpr = np.array([0.0, 1.0])
dummy_thresholds = np.array([np.inf, -np.inf])

# Get the FPR, TPR, and thresholds for the logistic model
model_fpr, model_tpr, thresholds = roc_curve(y_true, y_score)

# Plot setup
fig = plt.figure(figsize=(12, 8))
plt.plot(dummy_fpr, dummy_tpr, linestyle='--', label='Dummy Model')  # Diagonal line for random classifier
plt.plot(model_fpr, model_tpr, marker='.', label='Logistic Model')  # Your model's ROC curve

# Add annotations for each threshold value
ax = fig.add_subplot(111)
for xyz in zip(model_fpr, model_tpr, thresholds):
    ax.annotate(f'{np.round(xyz[2], 2)}', xy=(xyz[0], xyz[1]), xytext=(xyz[0] + 0.05, xyz[1] - 0.05),
                arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=10)

# Labels and legend
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()

# Show the plot
plt.show()
```
